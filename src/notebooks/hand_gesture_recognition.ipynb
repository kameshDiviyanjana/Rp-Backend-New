```python
# Import necessary libraries
import os
import matplotlib.pyplot as plt
from PIL import Image

# Define dataset directory
dataset_dir = '../dataset/train/'

# List all classes
classes = os.listdir(dataset_dir)
print(f"Classes: {classes}")

# Check number of images in each class
class_counts = {}
for cls in classes:
    class_path = os.path.join(dataset_dir, cls)
    class_counts[cls] = len(os.listdir(class_path))

print("Number of images per class:")
for cls, count in class_counts.items():
    print(f"Class {cls}: {count} images")

# Display a few sample images from each class
def display_sample_images(class_dir, num_images=3):
    plt.figure(figsize=(15, 5))
    for i, img_name in enumerate(os.listdir(class_dir)[:num_images]):
        img_path = os.path.join(class_dir, img_name)
        img = Image.open(img_path)
        plt.subplot(1, num_images, i+1)
        plt.imshow(img)
        plt.title(f"Sample {i+1}")
        plt.axis('off')
    plt.show()

# Display samples for each class
for cls in classes:
    print(f"Class: {cls}")
    class_dir = os.path.join(dataset_dir, cls)
    display_sample_images(class_dir)
```
